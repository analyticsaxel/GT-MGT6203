```r

temp <- Carseats %>% 
    mutate (Bad_Shelf = ifelse(Carseats$ShelveLoc == 'Bad', 1, 0) ) %>%
    mutate (Good_Shelf= ifelse(Carseats$ShelveLoc == 'Good', 1, 0))

temp
```

```r
lm.fit2=lm(temp$Sales~temp$Price+temp$Bad_Shelf+temp$Good_Shelf)
summary(lm.fit2)
```

```r
#The third part of the HW
PriceDemand = read.csv("PriceDemand.csv", header = TRUE)
```

```r
nrow(PriceDemand)
head(PriceDemand)
```

```r
lm.fit3 = lm(PriceDemand$Qty~PriceDemand$Price)
summary(lm.fit3)
```

```r
temp <- PriceDemand %>% 
    mutate(Price_ln = log(PriceDemand$Price, base = exp(1)))
```

```r
head(temp)
```

```r
lm.fit4 = lm(temp$Qty~temp$Price_ln)
summary(lm.fit4)
```

```r
temp <- temp %>% 
    mutate(Qty_ln = log(temp$Qty, base = exp(1)))
```

```r
head(temp)
```

```r
lm.fit5 = lm(temp$Qty_ln~temp$Price)
summary(lm.fit5)
```

```r
lm.fit6 = lm(temp$Qty_ln~temp$Price_ln)
summary(lm.fit6)
```
